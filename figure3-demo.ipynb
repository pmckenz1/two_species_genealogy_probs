# Recreating Figure 3 from Rosenberg 2003 

Using adapted equations 14, 15, 16, and 17 of Rosenberg, "The Shapes of Neutral Gene Genealogies in Two Species: Probabilities of Monophyly, Paraphyly, and Polyphyly in a Coalescent Model", Evolution 2003 https://doi.org/10.1111/j.0014-3820.2003.tb00355.x

### Load in the functions:

```r
source('./R/rosenberg2003_eqs.R')
```

### Demonstrate the functions:

We are given rA, rB, Ta, and Tb:

```r
ra = 9
rb = 25
ta = 1.3
tb = 2
```

Get the probability of **reciprocal monophyly** with `PrC1`:

```r
PrC1(ra = ra,
     rb = rb,
     coalT_A = ta
     ,coalT_B = tb)
```

Get the probability of **monophyletic A** and **paraphyletic B** with `PrC2`:

```r
PrC2(ra = ra,
     rb = rb,
     coalT_A = ta
     ,coalT_B = tb)
```

Get the probability of **paraphyletic A** and **monophyletic B** with `PrC3`:

```r
PrC3(ra = ra,
     rb = rb,
     coalT_A = ta
     ,coalT_B = tb)
```

Get the probability of **polyphyly** with `PrC4`:

```r
PrC4(ra = ra,
     rb = rb,
     coalT_A = ta
     ,coalT_B = tb)
```

You can see that they **add to 1** if the parameters stay the same across the four functions: 

```r
PrC1(ra = ra,
     rb = rb,
     coalT_A = ta
     ,coalT_B = tb) + 
PrC2(ra = ra,
     rb = rb,
     coalT_A = ta
     ,coalT_B = tb) + 
PrC3(ra = ra,
     rb = rb,
     coalT_A = ta
     ,coalT_B = tb) + 
PrC4(ra = ra,
     rb = rb,
     coalT_A = ta
     ,coalT_B = tb)
```

## Figure 3i

```r
# set parameter values
ra = 3
rb = 3

# make vectors of probabilities across time values
probsC1 <- integer(0)
probsC2 <- integer(0)
probsC3 <- integer(0)
probsC4 <- integer(0)
for (coal_time in seq(0,5,by=0.1)) {
  probsC1 <- c(probsC1,PrC1(ra,rb,coal_time, coal_time)) # coal times are equal
  probsC2 <- c(probsC2,PrC2(ra,rb,coal_time, coal_time))
  probsC3 <- c(probsC3,PrC3(ra,rb,coal_time, coal_time))
  probsC4 <- c(probsC4,PrC4(ra,rb,coal_time, coal_time))
}

plot(x=seq(0,5,by=0.1),
     y=probsC1,
     ylim=c(0,1),
     main='Figure 3i, ra = rb = 3, Tb = Ta',
     xlab = 'time Ta (coalescent units)',
     ylab = 'prob',
     pch = 5,
     type='o')
points(x=seq(0,5,by=0.1),y=probsC2,pch=3,type = 'o')
points(x=seq(0,5,by=0.1),y=probsC3,pch=0,type = 'o')
points(x=seq(0,5,by=0.1),y=probsC4, pch=4,type = 'o')
```

## Figure 3ii

```r
# set parameter values
ra = 3
rb = 3

# make vectors of probabilities across time values
probsC1 <- integer(0)
probsC2 <- integer(0)
probsC3 <- integer(0)
probsC4 <- integer(0)
for (coal_time in seq(0,5,by=0.1)) {
  probsC1 <- c(probsC1,PrC1(ra,rb,coal_time, coal_time/2)) # coal times are different!
  probsC2 <- c(probsC2,PrC2(ra,rb,coal_time, coal_time/2))
  probsC3 <- c(probsC3,PrC3(ra,rb,coal_time, coal_time/2))
  probsC4 <- c(probsC4,PrC4(ra,rb,coal_time, coal_time/2))
}

plot(x=seq(0,5,by=0.1),
     y=probsC1,
     ylim=c(0,1),
     main='Figure 3ii, ra = rb = 3, Tb = Ta/2',
     xlab = 'time Ta (coalescent units)',
     ylab = 'prob',
     pch = 5,
     type = 'o')
points(x=seq(0,5,by=0.1),y=probsC2,pch=3,type = 'o')
points(x=seq(0,5,by=0.1),y=probsC3,pch=0,type = 'o')
points(x=seq(0,5,by=0.1),y=probsC4, pch=4,type = 'o')
```
